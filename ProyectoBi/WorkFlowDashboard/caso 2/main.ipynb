```python
import pandas as pd
import numpy as np
from IPython.display import display as dp
```

```python
df= pd.read_csv('ventas.csv')
```

```python
dp(df)
```

```python
df=df.drop("id_venta", axis=1)
```

```python
dp(df)
```

```python
with pd.ExcelWriter("formateado.xlsx", engine="xlsxwriter") as writer:
    df.to_excel(writer, sheet_name="Hoja1", index=False)

    workbook  = writer.book
    worksheet = writer.sheets["Hoja1"]

    # Formatos
    formato_entero = workbook.add_format({"num_format": "0"})
    formato_float  = workbook.add_format({"num_format": "0.00"})
    formato_fecha  = workbook.add_format({"num_format": "yyyy-mm-dd"})

    # Aplicar formatos a columnas (columna A=0, B=1, etc.)
    worksheet.set_column("A:A", 10, formato_entero)  # Enteros
    worksheet.set_column("B:B", 12, formato_float)   # Flotantes
    worksheet.set_column("D:D", 15, formato_fecha)   # Fechas
```

```python
with pd.ExcelWriter("Ventas_formateadas.xlsx", engine="xlsxwriter") as ventas:
    df.to_excel(ventas, sheet_name="Ventas", index=False)

    libro = ventas.book
    hoja = ventas.sheets["Ventas"]

    fdate  = libro.add_format({"num_format": "yyyy-mm-dd"})

    hoja.set_column("A:A",15, fdate)

    
    
               
```
